```python
plt.figure(figsize=(10,7))

#Config o boxplot do alcool por qualidade
dados.boxplot(column='alcohol', by='quality')
# TÃ­tulos e rÃ³tulos
plt.title('DistribuiÃ§Ã£o do teor alcoolico por Qualidade do Vinho')
plt.suptitle('')  # Remove o tÃ­tulo automÃ¡tico do Pandas
plt.xlabel('Qualidade do Vinho')
plt.ylabel('alcohol')
plt.grid(True, alpha=0.3)

plt.show()
```

```python
media_alc = dados.groupby("quality")["alcohol"].mean()
mediana_alc = dados.groupby("quality")["alcohol"].median()
maximo_alc = dados.groupby("quality")["alcohol"].max()

stats = pd.DataFrame({
    "MÃ©dia": round(media_alc,2),
    "Mediana": round(mediana_alc,2),
    "MÃ¡ximo": maximo_alc
})

stats
```

Resposta:   
- A qualidade mÃ©dia aumenta junto com o teor alcoÃ³lico mÃ©dio.
- O vinho com qualidade 5 tem uma mÃ©dia de Ã¡lcool mais baixa, mas inclui o maior outlier (14.9%).
- Vinhos de qualidade 8 tÃªm os maiores valores mÃ©dios e mediana de Ã¡lcool.

```python
import seaborn as sns

# Tamanho do grÃ¡fico
plt.figure(figsize=(8, 6))

# GrÃ¡fico de dispersÃ£o com linha de regressÃ£o
sns.regplot(
    x='volatile acidity', 
    y='alcohol', 
    data=dados, 
    scatter_kws={'alpha': 0.5},  # transparÃªncia dos pontos
    line_kws={'color': 'red'}  # cor da linha
)

# TÃ­tulos e rÃ³tulos
plt.title('RelaÃ§Ã£o entre Acidez VolÃ¡til e Teor AlcoÃ³lico')
plt.xlabel('Acidez VolÃ¡til')
plt.ylabel('Teor AlcoÃ³lico (%)')
plt.grid(True)

# Exibir o grÃ¡fico
plt.tight_layout()
plt.show()
```

Vinhos de melhor qualidade costumam ter acidez volÃ¡til mais controlada e nÃ­vel de Ã¡lcool mais elevado.

Isso reforÃ§a as observaÃ§Ãµes anteriores: vinhos melhores tendem a ser mais alcoÃ³licos e menos Ã¡cidos (no sentido da acidez volÃ¡til, que estÃ¡ associada a vinhos com aroma avinagrado).

O valor Ã© negativo, o que confirma que hÃ¡ uma relaÃ§Ã£o inversa: quanto maior a acidez volÃ¡til, menor tende a ser o teor alcoÃ³lico.
