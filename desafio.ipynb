# Distribuição do teor alcóolico por qualidade do vinho

```python
dados[['quality','alcohol']].head()
```

```python
media_alcohol_qualidade = dados.groupby('quality')['alcohol'].mean()
media_alcohol_qualidade
```

```python
plt.figure(figsize=(10,7))

#Config o boxplot do alcool por qualidade
dados.boxplot(column='alcohol', by='quality')
# Títulos e rótulos
plt.title('Distribuição do teor alcoolico por Qualidade do Vinho')
plt.suptitle('')  # Remove o título automático do Pandas
plt.xlabel('Qualidade do Vinho')
plt.ylabel('alcohol')
plt.grid(True, alpha=0.3)

plt.show()
```

```python
media_alc = dados.groupby("quality")["alcohol"].mean()
mediana_alc = dados.groupby("quality")["alcohol"].median()
maximo_alc = dados.groupby("quality")["alcohol"].max()

stats = pd.DataFrame({
    "Média": round(media_alc,2),
    "Mediana": round(mediana_alc,2),
    "Máximo": maximo_alc
})

stats
```

Resposta:   
- A qualidade média aumenta junto com o teor alcoólico médio.
- O vinho com qualidade 5 tem uma média de álcool mais baixa, mas inclui o maior outlier (14.9%).
- Vinhos de qualidade 8 têm os maiores valores médios e mediana de álcool.

```python
import seaborn as sns

# Tamanho do gráfico
plt.figure(figsize=(8, 6))

# Gráfico de dispersão com linha de regressão
sns.regplot(
    x='volatile acidity', 
    y='alcohol', 
    data=dados, 
    scatter_kws={'alpha': 0.5},  # transparência dos pontos
    line_kws={'color': 'red'}  # cor da linha
)

# Títulos e rótulos
plt.title('Relação entre Acidez Volátil e Teor Alcoólico')
plt.xlabel('Acidez Volátil')
plt.ylabel('Teor Alcoólico (%)')
plt.grid(True)

# Exibir o gráfico
plt.tight_layout()
plt.show()
```

Vinhos de melhor qualidade costumam ter acidez volátil mais controlada e nível de álcool mais elevado.

Isso reforça as observações anteriores: vinhos melhores tendem a ser mais alcoólicos e menos ácidos (no sentido da acidez volátil, que está associada a vinhos com aroma avinagrado).

O valor é negativo, o que confirma que há uma relação inversa: quanto maior a acidez volátil, menor tende a ser o teor alcoólico.
